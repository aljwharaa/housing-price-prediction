```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), cmap='viridis')
```

```python
df.replace("?", np.nan, inplace = True)
df.head()
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=["lat"], axis=0, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(["yr_renovated"], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df['sqft_lot']=df['sqft_lot'].fillna(df['sqft_lot'].mean())
df['sqft_lot'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df['bathrooms'].value_counts()
```

```python
df['bathrooms']=df['bathrooms'].fillna(df['bathrooms'].value_counts().idxmax())

df['bathrooms'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df['Old'].fillna("yes", inplace = True)

df['Old'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imr = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
df['view'] = imr.fit_transform(df[['view']])

df['view'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df['waterfront']=df['waterfront'].fillna(df['waterfront'].value_counts().idxmax())

df['waterfront'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df['bedrooms'] = imr.fit_transform(df[['bedrooms']])
df['bedrooms'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.drop(['New'], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['sqft_living'].dtypes
```

```python
df['sqft_living'] = df['sqft_living'].astype("float")
df['sqft_above'] = df['sqft_above'].astype("float")
df['sqft_basement'] = df['sqft_basement'].astype("float")
```

```python
df.info()
```

```python
df
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df['Old'].unique()
```

```python
df.replace("TRUE", 'yes', inplace = True)
df.replace("FALSE", 'no', inplace = True)
df.replace("YES", 'yes', inplace = True)
df.replace("NO", 'no', inplace = True)
```

```python
df['Old'].unique()
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
df[['sqft_living','sqft_lot']] = min_max_scaler.fit_transform(df[['sqft_living','sqft_lot']])
```
